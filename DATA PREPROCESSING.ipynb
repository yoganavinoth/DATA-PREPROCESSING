```python
import pandas as pd
```

```python
dataset=pd.read_csv("kidney_disease.csv")
dataset
```

```python
dataset.isnull().sum()
```

```python
def quanQual(dataset):
    quan=[]
    qual=[]
    for columnName in dataset.columns:
    #print(columnName)
        if(dataset[columnName].dtypes=='O'):
        #print("qual")
            qual.append(columnName)
        else:
        #print("quan")
             quan.append(columnName)
    return quan,qual
```

```python
quan,qual=quanQual(dataset)
```

```python
quan
```

```python
qual
```

```python
dataset[quan]
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan,strategy="mean",verbose=0,copy=True)
imp.fit(dataset[quan])
datan=imp.transform(dataset[quan])
```

```python
imp
```

```python
datan
```

```python
datan=pd.DataFrame(datan,columns=quan)
```

```python
datan
```

```python
dataset[qual]
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan,strategy="most_frequent")
imp.fit(dataset[qual])
datal=imp.transform(dataset[qual])
```

```python
imp
```

```python
datal
```

```python
datal=pd.DataFrame(datal,columns=qual)
```

```python
datal
```

```python
dataset=pd.read_csv("kidney_disease.csv")
```

```python
dataset=pd.concat([datan,datal],axis=1)
```
